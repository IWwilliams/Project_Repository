```
def readCSV(filepath):

    df = (spark.read
            .option("multiline", "true")
            .option("header", "true")
            .option("escape", '"')
            .csv(filepath))
    return df
```

```
filepath = "dbfs:/FileStore/shared_uploads/hartbraydon@live.ca/xlsxYoutubeData_and_MovieMetaData-3.csv"
df = readCSV(filepath)
```

```
from pyspark.ml.feature import StringIndexer

studioEncoder = StringIndexer(inputCol = 'Primary Studio', outputCol = 'studio').fit(df)
df = studioEncoder.transform(df)
```

```
from pyspark.sql.types import IntegerType
from pyspark.sql.types import DoubleType
from pyspark.sql.functions import col
df = df.select(col("viewCount").cast(IntegerType()).alias("viewCount"), col("likeCount").cast(IntegerType()).alias("likeCount"), col("commentCount").cast(IntegerType()).alias("commentCount"), col("Budget").cast(IntegerType()).alias("Budget"), col("OpeningWeek Income").cast(IntegerType()).alias("label"), col("Action").cast(IntegerType()).alias("Action"), col("Adventure").cast(IntegerType()).alias("Adventure"), col("Drama").cast(IntegerType()).alias("Drama"), col("Animation").cast(IntegerType()).alias("Animation"), col("Fantasy").cast(IntegerType()).alias("Fantasy"), col("Comedy").cast(IntegerType()).alias("Comedy"), col("Sci-Fi").cast(IntegerType()).alias("Sci-Fi"), col("Family").cast(IntegerType()).alias("Family"), col("Crime").cast(IntegerType()).alias("Crime"), col("Thriller").cast(IntegerType()).alias("Thriller"), col("Horror").cast(IntegerType()).alias("Horror"), col("Mystery").cast(IntegerType()).alias("Mystery"), col("War").cast(IntegerType()).alias("War"), col("Biography").cast(IntegerType()).alias("Biography"), col("Romance").cast(IntegerType()).alias("Romance"), col("Musical").cast(IntegerType()).alias("Musical"), col("History").cast(IntegerType()).alias("History"), col("Sport").cast(IntegerType()).alias("Sport"), col("studio").cast(IntegerType()).alias("studio"), col("sentiment").cast(DoubleType()).alias("sentiment"), col("ReleaseMonth").cast(IntegerType()).alias("ReleaseMonth"))
display(df)
```

```
from pyspark.ml.feature import OneHotEncoder

encoder = OneHotEncoder(inputCol = "studio", outputCol = 'Studio_encoded', dropLast= False)
models = encoder.fit(df)
df = models.transform(df)

encoder = OneHotEncoder(inputCol = "ReleaseMonth", outputCol = 'ReleaseMonth_encoded', dropLast= False)
models = encoder.fit(df)
df = models.transform(df)
```

```
train, test = df.randomSplit([0.8,0.2],24)
```

```
from pyspark.ml.feature import MinMaxScaler
from pyspark.ml.linalg import Vectors
from pyspark.ml import Pipeline
from pyspark.ml.feature import VectorAssembler

columns = ["viewCount", "likeCount", "commentCount", "Budget"]

for i in columns:
    assembler = VectorAssembler(inputCols = [i], outputCol = i+"_Vect")
    train = assembler.transform(train)
    test = assembler.transform(test)
    scaler = MinMaxScaler(inputCol = i+"_Vect", outputCol = i + "_Scaled")
    scalerModel = scaler.fit(train.select(i+"_Vect"))
    train = scalerModel.transform(train)
    train = train.drop(i+"_Vect")
    test = scalerModel.transform(test)
    
```

```
important_features = ['Action', 'Adventure', 'Drama', 'Animation', 'Fantasy', 'Comedy', 'Sci-Fi', 'Family', 'Crime', 'Thriller', 'Horror', 'Mystery', 'War', 'Biography', 'Romance', 'Musical', 'History', 'Sport', 'studio', 'sentiment', 'Studio_encoded', 'viewCount_Scaled', 'likeCount_Scaled', 'commentCount_Scaled', 'Budget_Scaled', 'ReleaseMonth_encoded']
```

```
from pyspark.ml.feature import VectorAssembler

vec_assembler = VectorAssembler(inputCols = important_features, outputCol = 'features')
train = vec_assembler.transform(train)
test = vec_assembler.transform(test)
```

```
from pyspark.ml.regression import LinearRegression
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder, CrossValidatorModel, TrainValidationSplit
from pyspark.ml.evaluation import RegressionEvaluator

lr = LinearRegression(labelCol = 'label', featuresCol='features')

lrparamGrid = (ParamGridBuilder()
               .addGrid(lr.regParam, [0.1, 0.01])
               .addGrid(lr.elasticNetParam, [0.0, 0.5, 1.0])
               .addGrid(lr.maxIter, [1, 5, 10, 20, 50]).build())

lrevaluator = RegressionEvaluator(predictionCol = "prediction", labelCol ="label", metricName ="rmse")

lrcv = CrossValidator(estimator = lr, estimatorParamMaps = lrparamGrid, evaluator = lrevaluator, numFolds = 3)

lrcvModel = lrcv.fit(train)
print(lrcvModel)

lrcvSummary = lrcvModel.bestModel.summary

# print("Coefficient Standard Errors: " + str(lrcvSummary.coefficientStandardErrors))
# print("P Values: " + str(lrcvSummary.pValues)) # Last element is the intercept

# Use test set here so we can measure the accuracy of our model on new data
lrpredictions = lrcvModel.transform(test)

# cvModel uses the best model found from the Cross Validation
# Evaluate best model
print('RMSE:', lrevaluator.evaluate(lrpredictions))


```

```
def hyperparameter_getter(model_obj,cv_fold = 2.0):
 
    regParam = []
    elasticNetParam  = []
    maxIter = []
    
    ## Get metrics
 
    metrics = model_obj.avgMetrics
    assert type(metrics) is list
    assert len(metrics) > 0
 
    ## Get the paramMap element
 
    for x in range(len(model_obj._paramMap.keys())):
        if list(model_obj._paramMap.keys())[x].name=='estimatorParamMaps':
            param_map_key = list(model_obj._paramMap.keys())[x]
 
    params = model_obj._paramMap[param_map_key]
 
    for i in range(len(params)):
        for k in params[i].keys():
            # print(k)
            # print(params[i][k])
            if k.name =='elasticNetParam':
                elasticNetParam.append(params[i][k])
            if k.name =='regParam':
                regParam.append(params[i][k])
            if k.name =='maxIter':
                maxIter.append(params[i][k])
 
    results_df =  pd.DataFrame({'model':'linear_regression','metrics':metrics, 
             'maxIter': maxIter, 
             'elasticNetParam': elasticNetParam, 
             'regParam':regParam})
 
    # Because of [SPARK-16831][PYTHON] 
    # It only sums across folds, doesn't average
    spark_version = [int(x) for x in sc.version.split('.')]
 
    if spark_version[0] <= 2:
        if spark_version[1] < 1:
            results_df.metrics = 1.0*results_df['metrics'] / cv_fold
 
    return results_df
```
